##  The factorial function is given below as a code

```python
def factorial(number):
    factorial=1
    if (number==0 or number==1):
        print("factorial", factorial)
    else:
        while (number>=1):
            factorial*=number
            number-=1
        print("factorial is",factorial)
        
factorial(8)
```
